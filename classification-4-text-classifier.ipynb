```python
%matplotlib inline
import os
import pandas as pd
import numpy as np
```

## Get data

```python
df = pd.read_csv('speeches.csv')
df
```

```python
# Ensure 'Filename' is not 'N/A' and not NaN
df = df[(df['Filename'] != 'N/A') & (df['Filename'].notna())]

# Using .loc to ensure we are modifying the DataFrame safely
df.loc[:, 'filepath'] = 'speeches/' + df['Filename']
df.loc[:, 'file_exists'] = df['filepath'].apply(lambda x: os.path.isfile(x))

# Display the first few rows of the DataFrame
df.head()
```

## Filter Data

```python
df = df[df['Type of Speech'].isin(['State of the state','Both'])]
f"Dataset is {len(df)} speeches"
```

## Read Speeches

```python
def get_speeches(df):
    speeches = []
    for path in df['filepath']:
        with open(path) as f:
            text = f.read()
            speeches.append(text)
    return speeches

speeches = get_speeches(df)
```

## Turn Speeches Into Vectors!

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

## YOU CAN EDIT THESE
y_columns = ['Party', 'Trifecta']
BINARY=True
NGRAMS=(1,1)
MIN_DF=0.0

def ngram_vectorizer(n, binary, min_df):
    return TfidfVectorizer(
        stop_words='english', # 'english' if not custom list
        ngram_range=NGRAMS,
        binary=binary,
        min_df=min_df
    )
```

```python
vectorizer = ngram_vectorizer(NGRAMS, binary=BINARY, min_df=MIN_DF)
X = vectorizer.fit_transform(speeches)
y = np.array(df['Party'])
```

```python
# showing you the array!
print(X.shape)
X.toarray()
```

```python
# turning it back into a dataframe for visibility
word_vectors = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
df[['State', 'Governor', 'Party', 'Trifecta Status']].reset_index().\
    merge(word_vectors, left_index=True, right_index=True).head()
```

```python
print(y)
y = (y == 'R').astype('int')
y
```

## Train/Test Split

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3)
```

## Train a Classifier

```python
# Import the classifier from scikit-learn
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB

# define model
clf = LogisticRegression(C=1e9, solver='lbfgs', max_iter=4000) 
# clf = MultinomialNB(alpha=1.0e-10, class_prior=None, fit_prior=True)

# fit model on training data
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y, clf.predict(X))
pd.DataFrame(matrix, 
             columns = ['pred_Dem', 'pred_Rep'], 
             index=['Dem', 'Rep'])\
    .assign(Total=lambda x: x.sum(axis=1))\
    .sort_values(by='Total')[['pred_Rep', 'pred_Dem', 'Total']]\
    .T.assign(Total = lambda x: x.sum(axis=1)).T
```

```python
from sklearn.metrics import recall_score,accuracy_score, precision_score, f1_score

# accuracy is the proportion of correctly predicted labels out of all labels
print("accuracy_score", round(accuracy_score(y_test, clf.predict(X_test)),2))

# note: for precision and recall, we're interested in what 1 is (in this case, Republican)

# precision is the proportion of correctly predicted Republicans out of all predicted Republicans
print("precision_score", precision_score(y_test, clf.predict(X_test)).round(2))
# recall is the proportion of correctly predicted Republicans out of all actual Republicans
print("recall_score", recall_score(y_test, clf.predict(X_test)).round(2))
# f1 score is the harmonic mean of precision and recall
print("f1_score", f1_score(y_test, clf.predict(X_test)).round(2))
```

##  Cross Validation

```python
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
# 5-fold cross-validation
scoring = ['accuracy', 'precision', 'recall', 'f1']
scores = cross_validate(clf, X, y, scoring=scoring, cv=4)
pd.DataFrame(scores).round(2)
```

```python
cross_val_score(clf, X, y)
```
